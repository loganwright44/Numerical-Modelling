```python
import numpy as np
from numpy.linalg import norm
import matplotlib.pyplot as plt

# Define some constants
G = 6.67e-11

class Body:
    def __init__(self, r : np.array, v : np.array, M : float):
        self.r = r
        self.v = v
        self.M = M

class Bodies:
    def __init__(self, bodies):
        self.bodies = bodies
        self.numBodies = len(bodies)
        self.k = np.array([
            [body.r for body in bodies],
            [body.v for body in bodies]
        ])
        self.r = self.k[0]
        self.v = self.k[1]
    
    def __len__(self):
        return len(self.r)
    
    def __iter__(self):
        return iter(self.bodies)

class ForceTensor:
    def __init__(self, dimension : int):
        # Define the dimension of the rank-2 tensor and then construct it
        self.dim = dimension
        self.dr = [[0.0 for i in range(self.dim)] for j in range(self.dim)]
        self.dv = [[0.0 for i in range(self.dim)] for j in range(self.dim)]
        
        # Count the number of unique tensor components
        self.n_components = sum([i for i in range(self.dim)])
    
    def dvTensor(self, bodies : np.array):
        if len(bodies) != self.dim:
            self.dim = len(bodies)
            self.tensor = [[0.0 for i in range(self.dim)] for j in range(self.dim)]
        
        for i, body1 in enumerate(bodies):
            for j, body2 in enumerate(bodies):
                if i == j:
                    continue
                else:
                    self.dv[i][j] = G * body2.M * \
                        (body2.r - body1.r) / (norm(body2.r - body1.r) ** 3)
        
        return np.array([sum(a) for a in self.dv])
    
    def drTensor(self):
        return np.array([sum(V) for V in self.dr])

class DifferentialEQSolver:
    def __init__(self, diffEQ):
        self.f = diffEQ
    
    def RungeKutta4(self, bodies, forceTensor, t, tFinal, dt):
        X = 0
        Y = 1
        
        r = []
        for i in range(len(bodies)):
            r.append([[], []])
        
        while t < tFinal:
            k1 = f(bodies, bodies.k, forceTensor, dt)
            k2 = f(bodies, bodies.k + dt / 2 * k1, forceTensor, dt / 2)
            k3 = f(bodies, bodies.k + dt / 2 * k2, forceTensor, dt / 2)
            k4 = f(bodies, bodies.k + dt * k3, forceTensor, dt)
            
            bodies.k += dt / 6 * (k1 + 2 * (k2 + k3) + k4)
            
            for i in range(len(bodies)):
                r[i][X].append(bodies.k[0][i][X])
                r[i][Y].append(bodies.k[0][i][Y])
            
            t += dt
        
        return r

def f(bodies, k, forceTensor, dt):
    dv = forceTensor.dvTensor(bodies)
    dr = k[1] + dv * dt
    return np.array([dr, dv], float)
    
# Define sun's kinematic parameters
rS = np.array([0.0, 0.0], float)
vS = np.array([0.0, 0.0], float)
eS = 0.0
MS = 1.99e30

Sun = Body(rS, vS, MS)

# Define earth's kinematic parameters
rE = np.array([1.474e11, 0.0], float)
vE = np.array([0.0, 3e4], float)
eE = 1.671e-2
ME = 5.972e24

Earth = Body(rE, vE, ME)

# Define moon's kinematic parameters
rM = np.array([3.844e8, 0.0]) + rE
vM = np.array([0.0, 1.022e3]) + vE
eM = 5.49e-2
MM = 3.748e22

Moon = Body(rM, vM, MM)

bodies = Bodies([Sun, Earth, Moon])

forceTensor = ForceTensor(len(bodies))

solve = DifferentialEQSolver(f)
x = solve.RungeKutta4(bodies, forceTensor, 0.0, 100000000.0, 1e5)

plt.plot(x[0][0], x[0][1])
plt.plot(x[1][0], x[1][1])
plt.plot(x[2][0], x[2][1])
plt.legend(["sun", "earth", "moon"])
plt.show()
bodies.k[0][2][0]
```
